```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
# %matplotlib inline # Makes matplots right under the cell instead of a different window, for older models
import matplotlib
matplotlib.rcParams["figure.figsize"] = (20,10) # Sets default figure size for matplots

df1 = pd.read_csv("rentfaster.csv")
df1.head()
```

```python
df1 = df1.drop(["rentfaster_id", "address", "link", "availability_date"], axis=1)
df1
```

```python
df1.groupby("lease_term")["lease_term"].agg("count")
```

```python
df1 = df1.drop(["lease_term"], axis=1)
df1
```

```python
df2 = df1.drop(["latitude", "longitude"], axis="columns")
df2
```

```python
df2.isnull().sum()
```

```python
df2 = df2.dropna(subset="beds")
df2 = df2.dropna(subset="baths")
df2 = df2.dropna(subset="sq_feet")

df2["cats"] = df2["cats"].fillna("False")
df2["dogs"] = df2["dogs"].fillna("False")
df2["smoking"] = df2["smoking"].fillna("Non-Smoking")

df2.isnull().sum()
df2.head()
```

```python
df2["furnishing"] = df2["furnishing"].replace("Unfurnished, Negotiable", "Unfurnished")
df2 = df2[df2["furnishing"] != "Negotiable"]
print(df2.furnishing.unique())
```

```python
df2["furnishingb"] = df2["furnishing"].apply(lambda x: 1 if x == "Furnished" else 0)
df2["smokingb"] = df2["smoking"].apply(lambda x: 0 if x == "Non-Smoking" else 1)
df2["catsb"] = df2["furnishing"].apply(lambda x: 1 if x else 0)
df2["dogsb"] = df2["furnishing"].apply(lambda x: 1 if x else 0)
df2.drop(["furnishing", "smoking", "cats", "dogs"], axis=1)
df2 = df2[df2["beds"]!="none Beds"]

df3 = df2.copy()
# df3 = df3[df3["beds"] != "Studio"]
# df3["beds"] = df3["beds"].apply(lambda x: int(x.split()[0]))
print(df3.shape)

df3["sq_feet"] = pd.to_numeric(df3["sq_feet"], errors="coerce")
df2 = df3 [ df3.sq_feet/df3.beds >= 300 ]
```

```python
location_stats = df2.groupby("city")["city"].agg("count").sort_values(ascending=False)
location_stats_less_than_10 = location_stats[location_stats < 10]
df2["city"] = df2["city"].apply(lambda x: "other" if x in location_stats_less_than_10 else x)
df4 = df2.drop(["furnishing", "smoking", "cats", "dogs"], axis="columns")
types_to_remove = ["Loft", "Vacation Home", "Acreage", "Parking Spot", "Mobile"]
df5 = df4[ ~df3["type"].isin(types_to_remove)]
df5
```

```python
def remove_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby("city"):
        mean = np.mean(subdf.price)
        stddev = np.std(subdf.price)
        smaller_subdf = subdf[ (subdf["price"]>= mean - stddev) & (subdf["price"]<= mean + stddev) ]
        df_out = pd.concat([df_out, smaller_subdf], ignore_index=True)
    return df_out
```

```python
# Time to turn the categorical data into a bunch of columns containing ones and zeros

dummy = pd.get_dummies(df6["city"], dtype="int")
dummy2 = pd.get_dummies(df6["type"], dtype="int")
dummy3 = dummy2.drop("Room For Rent", axis=1)
df7 = pd.concat([df6,dummy.drop("other",axis=1),dummy3],axis=1)
df7.head()
```

```python
df8 = df7.drop(["city", "province","type"], axis=1)
df8.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

X = df8.drop(["price"],axis=1)
y = df8["price"]

#Use random_state=60 (best linear regression seed
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=60)
model = RandomForestRegressor(n_estimators=100, random_state=60)
model.fit(X_train, y_train)
model.score(X_test, y_test)



# from sklearn.model_selection import train_test_split
# from sklearn.linear_model import LinearRegression

# X = df8.drop(["price"], axis=1)
# y = df8["price"]

# max = 0
# maxi = 0

# for i in range(100):
#     X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=i)

#     linRegModel = LinearRegression()
#     linRegModel.fit(X_train, y_train)
    
#     score = linRegModel.score(X_test, y_test)
    
#     if score > max:
#         max = score
#         maxi = i

# print(max)
# print(maxi)
```
